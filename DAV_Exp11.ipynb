```python
 import pandas as pd
```

```python
df_ENE = pd.read_excel("/content/Yearly Demand Profile_E_NE.xlsx")
df_NS = pd.read_excel("/content/Yearly Demand Profile_N_S.xlsx")
df_WS = pd.read_excel("/content/Yearly Demand Profile_W_S.xlsx")
```

```python
df_ENE.head()
```

```python
Yearly_demand_df = pd.concat([df_ENE, df_NS, df_WS], ignore_index=True)
Yearly_demand_df.head()
```

```python
Yearly_demand_df.shape
```

```python
Yearly_demand_df['Date'] = '2022 ' + Yearly_demand_df['Date']
Yearly_demand_df['Date'] = pd.to_datetime(Yearly_demand_df['Date'], format='%Y %d-%b %I%p')
Yearly_demand_df.head()
```

```python
import folium
```

```python
def get_coordinates(region):
    region_coordinates = {
      'Eastern Region - 2022': (23.684994, 86.952393),
      'Western Region - 2022': (22.814, 72.507),
      'Northern Region - 2022': (29.0588, 76.0856),
      'Southern Region - 2022': (11.1271, 78.6569),
      'North-Eastern Region - 2022': (26.2006, 92.9376)
    }
    return region_coordinates.get(region, (0, 0))
```

```python
india_map = folium.Map(location=[20.5937, 78.9629], zoom_start=5)
```

```python
for index, row in Yearly_demand_df.iterrows():
    folium.Marker(
        location=get_coordinates(row['Region']),
        popup=f"Region: {row['Region']}\nDate: {row['Date']}\nHourly Demand: {row['Hourly Demand Met (in MW)']}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(india_map)

# Display the map
india_map.save('energy_demand_map.html')
```
